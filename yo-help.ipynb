```
from pyspark.sql import functions as F
from pyspark.sql.window import Window
from pyspark.sql.types import *
from pyspark.storagelevel import StorageLevel
from pyspark.sql.functions import udf
from pyspark.sql.types import *
```

```
list=[['2018-01-01 11:00:00','10srinu200'],['2018-01-01 12:10:00','10drino200']]
cSchema = StructType([StructField("Date", StringType())\
                      ,StructField("Sesh", StringType())])
df= spark.createDataFrame(list,schema=cSchema)
display(df)
```

```
def userdefined(name):
    ans=[]
    for i in name[2:7]:
        ans.append(name[0:2]+i+name[7:10])
    return ans


udf1= udf(userdefined, ArrayType(StringType()))
df1= df.withColumn("Names", udf1('Sesh'))
display(df1)
```

```
df2= df1.withColumn("Names2",F.explode(df1.Names))
display(df2)
```
